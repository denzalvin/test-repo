```python
import pandas as pd
import numpy as np
from google.colab import files
import io


```

Upload the cvs file and covert it to a dataframe

```python
uploaded = files.upload()
data = io.BytesIO(uploaded['sample.csv']) 
wf = pd.read_csv(data )
```

Function that Clean up the address

```python
def add_clean(address):
  import re
  address = re.sub("[^a-zA-Z0-9 ]+","",address)
  address =address.lower()
  
  

  toReplace = {"st": 'Street', "rd": "Road", "ave": "Avenue","dr":"Drive","blvd":"BOULEVARD"}
  for k,v in toReplace.items():
    address = re.sub(r"\b" + k + r"\b", v, address)

  address=address.strip().title()
  
  return address
```

Using the Lambda and apply to apply to append to data frame

```python
wf['ouput']= wf['address'].apply(lambda x: add_clean(x))

wf
```
